```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('data.csv')
plt.figure(figsize=(15, 8))
# Group by 'Competition' and sum the goals for each competition
goals_by_competition = df.groupby('Competition')['Result'].apply(lambda x: x.apply(lambda y: int(y.split(':')[0])).sum()).reset_index()


# Plot the bar graph
plt.bar(goals_by_competition['Competition'], goals_by_competition['Result'])
plt.xlabel('Competition')
plt.ylabel('Number of Goals')
plt.title('Number of Goals Scored in Each Competition')
plt.show()
```

```python
# Group by 'Competition' and sum the goals for each competition
goals_by_competition = df.groupby('Competition')['Result'].apply(lambda x: x.apply(lambda y: int(y.split(':')[0])).sum()).reset_index()
plt.figure(figsize=(15, 8))
# Plot the pie chart
plt.pie(goals_by_competition['Result'], labels=goals_by_competition['Competition'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Distribution of Goals Across Competitions')
plt.show()
```

```python
goals_by_competition = df.groupby('Competition')['Result'].apply(lambda x: x.apply(lambda y: int(y.split(':')[0])).sum()).reset_index()
plt.figure(figsize=(15, 8))
# Plot the histogram
plt.bar(goals_by_competition['Competition'], goals_by_competition['Result'], color='skyblue')
plt.xlabel('Competition')
plt.ylabel('Number of Goals')
plt.title('Distribution of Goals Across Competitions')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
plt.figure(figsize=(15, 8))
# Group by 'Competition' and extract goals for each competition
goals_by_competition = df.groupby('Competition')['Result'].apply(lambda x: x.apply(lambda y: int(y.split(':')[0])).tolist()).reset_index()

# Plot the box plot
plt.boxplot(goals_by_competition['Result'], labels=goals_by_competition['Competition'])
plt.xlabel('Competition')
plt.ylabel('Number of Goals')
plt.title('Distribution of Goals Across Competitions')
plt.show()
```
